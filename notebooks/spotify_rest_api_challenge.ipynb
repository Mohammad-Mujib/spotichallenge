```python
base_url = 'https://api.spotify.com/v1/'
resource = 'artists/'
artist_id = '4Y7tXHSEejGu1vQ9bwDdXW'
related = '/related-artists'
print(base_url + resource + artist_id + related)
```

```python
full_endpoint = requests.get(base_url + resource + artist_id + related,  headers = header_info).json()
full_endpoint
```

```python
#top-tracks
base_url = 'https://api.spotify.com/v1/'
resource = 'artists/'
artist_id = '4Y7tXHSEejGu1vQ9bwDdXW'
country = '?market=es'
related = '/top-tracks'

full_endpoint_2 = requests.get(base_url + resource + artist_id + related + country,  headers = header_info).json()
full_endpoint_2
```

```python
full_endpoint_2 = base_url + resource + '{}' + related + country
print(full_endpoint_2)

for artist in artists_list:
    response = requests.get(full_endpoint_2.format(artist), headers=header_info)
    print([response.json()['tracks'][track]['name'] for track in range(len(response.json()['tracks']))])
```
